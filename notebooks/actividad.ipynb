```python
import pandas as pd
import numpy as np
import re
```

```python
url ="https://raw.githubusercontent.com/oswram19/etl-data-pipline-1764382012/refs/heads/main/data/G_actividad.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.shape
df.columns
df.info()
df.isnull().sum()
```

LIMPIAR DATOS

```python
df['fecha_actividad'] = pd.to_datetime(df['fecha_actividad'], errors='coerce')
display(df.head())
```

```python
df['id_usuario'].fillna('Unknown', inplace=True)
df.dropna(subset=['fecha_actividad'], inplace=True)

print('Missing values after cleaning:')
display(df.isnull().sum())

print('DataFrame information after cleaning:')
display(df.info())
```

```python
df_encoded = pd.get_dummies(df, columns=['accion', 'modulo'], prefix=['accion', 'modulo'])

display(df_encoded.head())

print('\nDimensiones del DataFrame Curado:', df_encoded.shape)
print('Nnueva columna creada:', [col for col in df_encoded.columns if col.startswith('accion_') or col.startswith('modulo_')])
```

```python
print('Información del DataFrame después de la codificación one-hot:')
display(df_encoded.info())

print('Valores faltantes después de la codificación one-hot:')
display(df_encoded.isnull().sum())
```

Separando Datos Curados y Rechazados

```python
# Regenerar df_encoded a partir del 'df' actual para asegurar la consistencia
# El 'df' actual ya ha manejado los valores faltantes en 'fecha_actividad' e 'id_usuario'
df_encoded = pd.get_dummies(df, columns=['accion', 'modulo'], prefix=['accion', 'modulo'])

print('Dimensiones del DataFrame Curado (regenerado):', df_encoded.shape)
display(df_encoded.head())
```

```python
# Cargar el CSV original para identificar las filas rechazadas
original_df = pd.read_csv(url)

# Convertir 'fecha_actividad' con errors='coerce' para identificar fechas inválidas
original_df['fecha_actividad_temp'] = pd.to_datetime(original_df['fecha_actividad'], errors='coerce')

# Identificar filas donde 'fecha_actividad' era inválida (filas rechazadas)
rejected_df = original_df[original_df['fecha_actividad_temp'].isna()].drop(columns=['fecha_actividad_temp'])

print('\nDimensiones de los Datos Rechazados:', rejected_df.shape)
display(rejected_df.head())

# Eliminar la columna temporal de original_df si no se necesita más tarde
del original_df['fecha_actividad_temp']
```

```python
# Guardar los datos curados
df_encoded.to_csv('actividad_curated.csv', index=False)
print('Datos curados guardados en curated_data.csv')

# Guardar los datos rechazados
rejected_df.to_csv('actividad_rejected.csv', index=False)
print('Datos rechazados guardados en rejected_data.csv')

print('\nSeparación y guardado completados!')
```
